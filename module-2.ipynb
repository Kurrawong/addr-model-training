## 1. QALI demo

QALI is Queensland's implementation of an application dedicated to the management of Address, Geographicla (Place) Names and Road Names data created according to the Address Model and sibling ICSM models.

KurrawognAI has a public-facing instance of Address Manager, the tool used for QALI at:

* <https://address-demo.dev.kurrawong.ai/>

You will need a username and passworkd from KurrawongAI to use this.

It is loaded with 1,000 real addresses from Queensland and testing addresses from the ACT.

We will walk through all parts of a test version of Address Manager as deployed for Queensland in QALI.

## 4. QLD Design

### Non-model elements

Data model elements other than those in the model that have been implemented for Qld are:

* QLD-specific vocabulary elements - anticipated by the model
* "Additional Properties" - shelving Qld data elements until the new address system is in stable production and these can be handled
    * these are easy to model in RDF!
    * we can see some of these in QALI
* We do expect that model elements will be almost all that is needed by Qld eventually

### Model feedback

Things that have changed in the model as a result of Qld's implementation:

* minimal - because the model was made within the Qld work
* address relations have been moved "out" of the Address Model as the relations can be calculated from Addressable Object relations
    * this needs further explaining in the Model's _Patterns_ section
