### Testing data

The online copy of Address Manager is loaded with 1,000 addresses from Queensland and the scenario testing data from ACT. More data will be added from jurisdictions as modelling tests are conducted.

Here are some queries that can be run on the test data:

```sparql
# count all addresses

PREFIX addr: <https://linked.data.gov.au/def/addr/>

SELECT (COUNT(?addr) AS ?count)
WHERE {
	?addr a addr:Address
}
```

```sparql
# gets 10 address IRIs, part types and values

PREFIX schema: <https://schema.org/>
PREFIX addr: <https://linked.data.gov.au/def/addr/>

SELECT ?addr ?at ?v
WHERE {
	?addr
        a addr:Address ;
    	schema:hasPart ?p ;
    .

	?p
        schema:value ?v ;
    	schema:additionalType ?at ;
    .
}
LIMIT 10
```

```sparql
# gets all facts about - an address

DESCRIBE <https://linked.data.gov.au/dataset/qld-addr/address/45fb0876-cdde-51f5-a94c-c1db2c731fde>
```
